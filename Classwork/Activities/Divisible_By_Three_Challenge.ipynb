# Your Mission üöÄ

Your task is to write a program that will do the following:

1. **Ask the User**: First, your program needs to ask the user how many numbers they want to check.  
2. **Looping Through Numbers**: Using a `for` loop, you will repeatedly:  
   * Prompt the user to enter a number.  
   * Tell them if that number is divisible by 3 or not.  
3. **Keep Count**: After the loop finishes, your program should display a summary:  
   * How many numbers entered were divisible by 3\.  
   * How many numbers entered were NOT divisible by 3\.

---

## Helpful Hints\! ‚ú®

* **Divisibility by 3**: A number is divisible by 3 if, when you divide it by 3, there's no remainder\! You'll want to use the **modulus (%) operator** for this. It gives you the remainder of a division. For example, `10 % 3` is `1`.  
* **Counting Time**: You'll need two separate counter variables:  
  * One to keep track of numbers divisible by 3\.  
  * One to keep track of numbers not divisible by 3\.
